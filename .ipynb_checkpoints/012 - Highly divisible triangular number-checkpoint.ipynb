```python
# The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 
#1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

# 1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

# Let us list the factors of the first seven triangle numbers:

# 1: 1
# 3: 1,3
# 6: 1,2,3,6
# 10: 1,2,5,10
# 15: 1,3,5,15
# 21: 1,3,7,21
# 28: 1,2,4,7,14,28
# We can see that 28 is the first triangle number to have over five divisors.

# What is the value of the first triangle number to have over five hundred divisors?
```

```python
def FindTriangleNumbers(limit):
    return sum(range(1, limit+1, 1))
```

```python
FindTriangleNumbers(10)
```

```python
def FindFactors(number):
    factors = []
    for i in range(1, number+1, 1):
        if number % i == 0:
            factors.append(i)
    return factors
```

```python
FindFactors(FindTriangleNumbers(7))
```

```python
def FindNumFactors(number):
    len = 0
    for i in range(1, number+1, 1):
        if number % i == 0:
            len += 1
    return len
```

```python
FindNumFactors(100000000)
```

```python
import numpy as np
```

```python
def FindPrimeNumbers(limit):
    sqrt_list = list(range(2, int(np.floor(np.sqrt(limit)))+1, 1))
    
    for step, elt in enumerate(sqrt_list):
        temp_list = []
        temp_list = [elt*elt+i*elt for i, elt in enumerate([elt]*limit) if elt*elt+i*elt<=limit]
        if step == 0:    
            PrimeNumbersList = [x for x in list(range(2, limit+1, 1)) if x not in temp_list]#[x for x in list(range(2, limit+1, 1)) if x not in temp_list]
        else:
            PrimeNumbersList = [x for x in PrimeNumbersList if x not in temp_list]
            
    return PrimeNumbersList
```

```python
def testFindNumFactors(number):
    primes = FindPrimeNumbers(number)
    
    len = 1
    
    for prime in primes:
        print('1: ', prime)
        while number / prime != 1:
            print('2: ', number, prime, number / prime)
            if number % prime == 0:
                len += 1
                number = number / prime
            else:
                continue
    
    return len
```

```python
testFindNumFactors(16)
```

```python
def Factor(n):
    Ans = []
    d = 2
    while d * d <= n:
        if n % d == 0:
            Ans.append(d)
            n //= d
        else:
            d += 1
    if n > 1:
        Ans.append(n)
    return Ans
```

```python
Factor(3472)
```
