```python
# The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 
#1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

# 1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

# Let us list the factors of the first seven triangle numbers:

# 1: 1
# 3: 1,3
# 6: 1,2,3,6
# 10: 1,2,5,10
# 15: 1,3,5,15
# 21: 1,3,7,21
# 28: 1,2,4,7,14,28
# We can see that 28 is the first triangle number to have over five divisors.

# What is the value of the first triangle number to have over five hundred divisors?
```

```python
def FindTriangleNumbers(limit):
    return sum(range(1, limit+1, 1))
```

```python
FindTriangleNumbers(10)
```

```python
def FindFactors(number):
    factors = []
    for i in range(1, number+1, 1):
        if number % i == 0:
            factors.append(i)
    return factors
```

```python
FindFactors(FindTriangleNumbers(7))
```

```python
def FindNumFactors(number):
    len = 0
    for i in range(1, number+1, 1):
        if number % i == 0:
            len += 1
    return len
```

```python
FindNumFactors(100000000)
```

```python
def FindPrimeNumbers(limit):
    a = list(range(limit+1))
    a[1] = 0
    lst = []

    i = 2
    while i <= limit:
        #print(a, lst)
        if a[i] != 0:
            lst.append(a[i])
            for j in range(i, limit+1, i):
                a[j] = 0
        i += 1
    return len(lst)+1
```

```python
FindPrimeNumbers(10000)
```
