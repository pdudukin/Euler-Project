```python
# Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

# Triangle Tn=n(n+1)/2  1, 3, 6, 10, 15, ...
# Pentagonal Pn=n(3n−1)/2  1, 5, 12, 22, 35, ...
# Hexagonal Hn=n(2n−1)  1, 6, 15, 28, 45, ...
# It can be verified that T285 = P165 = H143 = 40755.

# Find the next triangle number that is also pentagonal and hexagonal.
```

```python
import tqdm
from tqdm import tqdm_notebook
```

```python
def Tri(n):
    return n*(n+1)/2

def Pent(n):
    return n*(3*n-1)/2

def Hex(n):
    return n*(2*n-1)
```

```python
Tri_list = [int(Tri(x)) for x in range(1,100001)]
Pent_list = [int(Pent(x)) for x in range(1,100001)]
Hex_list = [int(Hex(x)) for x in range(1,100001)]
```

```python
res = []

for tr in tqdm.tqdm(Tri_list, total = len(Tri_list), desc = "Searching:"):
    if tr in Pent_list and tr in Hex_list:
        res.append(tr)
    
    if len(res)==3:
        break

res[2]
```
